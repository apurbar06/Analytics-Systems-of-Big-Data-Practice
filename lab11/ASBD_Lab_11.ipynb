### 1. For the given input file, calculate Wordcount using Hadoop MapReduce(optional) and Spark. Also, develop an equivalent conventional program (without spark RDDs) and compare the time taken by the two versions (one with spark and other without spark).

```python
import time


# Read the input file and split it into words
lines = sc.textFile("input.txt")
words = lines.flatMap(lambda line: line.split())

# Count the occurrences of each word
start_time = time.time()
word_count = words.map(lambda word: (word, 1)).reduceByKey(lambda x, y: x + y)
end_time = time.time()

# Print the results
for word, count in word_count.collect():
    print(f"{word}: {count}")

# Print the total time taken
print(f"\n\nTotal time taken: {end_time - start_time:.2f} seconds")
```

```python
import time

# Read the input file and split it into words
with open("input.txt", "r") as f:
    text = f.read()
words = text.split()

# Count the occurrences of each word
word_count = {}
start_time = time.time()
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
end_time = time.time()

# Print the results
for word, count in word_count.items():
    print(f"{word}: {count}")

# Print the total time taken
print(f"Total time taken: {end_time - start_time:.2f} seconds")
```

### 2. Randomly populate 100k numbers and calculate mean, variance, standard deviation for the generated data.

```python
import random

# Generate 100k random numbers
nums = [random.randint(1, 1000) for _ in range(100000)]

# Create an RDD of the numbers
nums_rdd = sc.parallelize(nums)

# Calculate the mean
mean = nums_rdd.mean()

# Calculate the variance
variance = nums_rdd.variance()

# Calculate the standard deviation
std_dev = nums_rdd.stdev()

# Print the results
print(f"Mean: {mean:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
```

### 3. Compute correlation between the given two series using Pearson’s and Spearman’s Method.
a. (Use the Spark MLlib libraries and helper functions available) </br>
i. Series A: 35, 23, 47, 17, 10, 43, 9, 6, 28 </br>
ii. Series B: 30, 33, 45, 23, 8, 49, 12, 4, 31

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
from pyspark.sql import SparkSession

# Initialize a SparkSession
spark = SparkSession.builder.appName("Correlation").getOrCreate()

# Create the input data
data = [(35, 30), (23, 33), (47, 45), (17, 23), (10, 8), (43, 49), (9, 12), (6, 4), (28, 31)]
df = spark.createDataFrame(data, ["A", "B"])

# Assemble the data into a feature vector
assembler = VectorAssembler(inputCols=["A", "B"], outputCol="features")
df_assembled = assembler.transform(df).select("features")

# Compute the correlation matrix using Pearson's method
pearson_matrix = Correlation.corr(df_assembled, "features", method="pearson")
pearson_correlation_coefficient = pearson_matrix.collect()[0]["pearson(features)"]
print(f"Pearson's correlation coefficient: \n {pearson_correlation_coefficient}")

# Compute the correlation matrix using Spearman's method
spearman_matrix = Correlation.corr(df_assembled, "features", method="spearman")
spearman_correlation_coefficient = spearman_matrix.collect()[0]["spearman(features)"]
print(f"\nSpearman's correlation coefficient: \n {spearman_correlation_coefficient}")

# Stop the SparkSession
spark.stop()
```

### 4. Randomly generate 100k numbers and apply the following functions on the generated random numbers. Develop two versions where in version-1 you will print all 100k numbers in the action statement whereas in the second version, print only 100 numbers using the action statement and compare the time taken by the two versions. </br>
a. Exponential Function, f1(x)=ex </br>
b. Logarithmic Function, f2(x)=log(x)

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, rand
import time
from pyspark.sql.functions import exp

spark = SparkSession.builder.appName("random_numbers").getOrCreate()
df = spark.range(100000).select(rand(seed=22).alias("random_number"))

# Print all 100k numbers
start_time_v1 = time.time()
df.withColumn("exp_number", exp(col("random_number"))).show(100000)
end_time_v1 = time.time()

# Print only 100 numbers
start_time_v2 = time.time()
df.withColumn("exp_number", exp(col("random_number"))).show(100)
end_time_v2 = time.time()
print("Time taken for version 1:", end_time_v1 - start_time_v1)
print("Time taken for version 2:", end_time_v2 - start_time_v2)

spark.stop()
```

### 5. Generate FIMs using FPgrowth algorithm on pyspark setup using a benchmark dataset available on the FIMI website.

```python
from pyspark.ml.fpm import FPGrowth
from pyspark.sql.functions import split

spark = SparkSession.builder.appName("FPGrowth").getOrCreate()
data = spark.read.csv("chess.dat", header=False, inferSchema=True)
data = data.selectExpr("_c0 as items")
data = data.select(split(data.items, " ").alias("items"))

fpGrowth = FPGrowth(minSupport=0.5, minConfidence=0.6)
model = fpGrowth.fit(data)

model.freqItemsets.show()
```
